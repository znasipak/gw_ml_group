# Importing gravitational wave time series data 

## Load necessary python libraries
While pandas is note necessary for loading the data, it is helpful for presenting the data contained in the various data files. If you do not want to use pandas, just skip to Part II

```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

## Part I: Properties of the sources in our gravitational wave data set

In this data set, gravitational wave sources are specified by 5 different parameters:
- $\mu/M_\odot$: the mass of the smaller black hole in solar masses $M_\odot$
- $M/M_\odot$: the mass of the supermassive black hole in solar masses $M_\odot$
- $a$: the dimensionless spin of the supermassive black hole
- $e_0$: the initial eccentricity of the black hole orbits when the binary is first observed
- $Y_0$: the initial inclination parameter of the black hole orbits when the binary is first observed

For this first data set, I've kept several external parameters constant. For example, the orientation of binary, its location in the sky, and its distance are all kept constant. Maybe these parameters can be varied for future data sets.

Below, I load a data frame that tells you which black hole systems exist in the data set. They are described by their filename and the 5 source parameters listed above, along with the spacing $dt$ of the time sampling. The time sampling is expressed in seconds.

```python
gw_dir = 'data/'
flist = os.listdir(gw_dir)
sources = pd.DataFrame(columns = ['a', 'e0', 'Y0', 'mu', 'M', 'dt'])
for f in flist:
    if f[-4:] == '.npz':
        params = np.load(gw_dir + f)
        s = pd.Series(params['params'].astype(float), index = ['a', 'e0', 'Y0', 'mu', 'M', 'dt'])
        df = s.to_frame().T.rename(index = {0:f})
        sources = sources.append(df)
```

We can load the parameter data frames by calling 'sources'

```python
sources
```

We can similarly sort the data according to one of the parameters

```python
sources.sort_values(by = 'a')
```

We can also see what are the unique parameter values used in our data set

```python
print('a values include ' + str(sorted(sources['a'].unique())))
print('e0 values include ' + str(sorted(sources['e0'].unique())))
print('Y0 values include ' + str(sorted(sources['Y0'].unique())))
print('mu values include ' + str(sorted(sources['mu'].unique())))
print('M values include ' + str(sorted(sources['M'].unique())))
```

We can then find all sources with a particular value of the black hole spin $a$

```python
sources[sources['a'] == 0.1]
```

## Part II: Load signal time series for a specific gravitational wave source

First we need to specify the parameters of the source

```python
mu = 30
M = 1e6
a = 0.9
e0 = 0.3
Y0 = 0.95
```

From this we can define the filename based on source properties

```python
filename = 'gw_a' + str(int(100*a)) + '_e' + str(int(100*e0)) + '_Y' + str(int(100*Y0)) + '_m' + str(int(mu)) + '_logM' + str(int(np.log10(M)))
```

Alternatively, we check that this source exists in our database by comparing to our sources data frame. If it exists, it returns the filename

```python
index = sources[(sources['mu'] == mu) & 
                   (sources['M'] == M) & 
                   (sources['a'] == a) & 
                   (sources['e0'] == e0) & 
                   (sources['Y0'] == Y0)].index
if index.size == 0:
    print('ERROR: Source does not exist')
else:
    filename = index[0]
    print("Source located in file " + filename)
```

Now we can load in our gravitational wave time series. Two series are actually given, with each series reperesenting a different (orthogonal) projection of the gravitational wave signal onto the LISA antenna (in the long wavelength approximation). The two projections are referred to as $h_I$ and $h_{II}$

```python
if os.path.exists(gw_dir + filename):
    gwdata = np.load(gw_dir + filename)
    hI = gwdata['hI']
    hII = gwdata['hII']
    t = np.arange(hI.size)*gwdata['params'][-1] # generate time data based on sampling dt
else:
    print('ERROR: Source does not exist')
```

Below we plot the two projections of the GW signal

```python
plt.plot(t, hI, label = '$h_I$')
plt.plot(t, hII, label = '$h_{II}$')
plt.xlabel('$t$ (seconds)')
plt.ylabel('GW strain $h(t)$')
plt.title('Two projections of the GW signal')
plt.legend(loc = 'lower left')
plt.show()
```
